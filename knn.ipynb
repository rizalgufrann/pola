```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(r'mobile_price.csv')
dataset
```

```python
df = dataset
df
```

```python
x = df[['battery_power', 'four_g', 'int_memory', 'ram']]
x
```

```python
y = df['price_range']
y
```

```python
#Import statement "from sklearn.model_selection import train_test_split" digunakan untuk mengimpor fungsi train_test_split dari modul model_selection dalam pustaka scikit-learn (sklearn).
from sklearn.model_selection import train_test_split
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

```python
#Melakukan penskalaan fitur menggunakan StandardScaler dari pustaka scikit-learn (sklearn) dalam persiapan data sebelum melatih dan menguji model.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train, y_train)
```

```python
x_test
```

```python
y_pred = knn.predict(x_test)
y_pred
```

```python
knn.predict_proba(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_pred = knn.predict(x_test)
print(classification_report(y_test, y_pred))
print("Accuracy:",accuracy_score(y_pred, y_test)*100,"%")
```
